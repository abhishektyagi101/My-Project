```python
# Creating target variables
dfy = pd.DataFrame(boston.target, columns=['target'])
```

```python
dfy.shape
```

```python
dfx.shape
```

```python
df_combine = pd.concat([dfx, dfy], axis=1)
```

```python
df_combine.shape
```

```python
df_combine.columns
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(df_combine.corr(), annot=True)
```

```python
fig,axs=plt.subplots(1,3,sharey=True)
df_combine.plot(kind='scatter', x='AGE', y='target', ax=axs[0], figsize=(15,10))
plt.title('Distribution')
df_combine.plot(kind='scatter', x='RM', y='target', ax=axs[1])
df_combine.plot(kind='scatter', x='LSTAT', y='target', ax=axs[2])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
x_train, x_test, y_train, y_test = train_test_split(dfx, dfy, test_size=0.3, random_state=42)
lr = LinearRegression()
lr.fit(x_train, y_train)
```

```python
y_predict=lr.predict(x_test)
```

```python
print(y_predict)
```

```python
from sklearn.metrics import r2_score, mean_squared_error
```

```python
mean_squared_error(y_test,y_predict)
```

```python
r2_score(y_test, y_predict)
```
