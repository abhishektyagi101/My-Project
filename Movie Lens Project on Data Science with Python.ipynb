```python
Data500 = Master_Data.iloc[:500, [0,7,8,4]]
```

```python
Data500.head()
```

```python
inputfeatures = Data500.iloc[:, :-1].values
label = Data500.iloc[:, -1].values
```

```python
inputfeatures.shape
```

```python
# Creating test train split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(inputfeatures, label, test_size=0.25, random_state=1)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
predictor  = KNeighborsClassifier(n_neighbors= 5)
predictor.fit(x_train, y_train)
```

```python
print(predictor.score(x_train, y_train))
```

```python
print(predictor.score(x_test,y_test))
```

```python
from sklearn.metrics import accuracy_score
predictions=predictor.predict(x_train)
print(accuracy_score(y_train, predictions))
```

```python
from sklearn.svm import SVC
predictor2 = SVC(gamma= "auto")
predictor2.fit(x_train, y_train)
print(predictor2.score(x_train, y_train))
print(predictor2.score(x_test, y_test))
```
